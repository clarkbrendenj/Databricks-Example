```python
%python
from pyspark.sql.functions import when

df = spark.table("samples.accuweather.forecast_daily_calendar_imperial")

# Select relevant columns and add rain_cat
df_selected = df.select(
    "city_name",
    "date",
    # Rain columns
    "has_rain",
    "rain_lwe_rate_avg",
    "rain_lwe_rate_max",
    "rain_lwe_rate_min",
    "rain_lwe_total",
    "rain_probability",
    "minutes_of_rain_total",
    # Wind columns
    "wind_direction_avg",
    "wind_gust_avg",
    "wind_gust_max",
    "wind_gust_min",
    "wind_gust_direction_avg",
    "wind_speed_avg",
    "wind_speed_max",
    "wind_speed_min",
    # Temperature columns
    "temperature_avg",
    "temperature_max",
    "temperature_min",
    "temperature_dew_point_avg",
    "temperature_dew_point_max",
    "temperature_dew_point_min",
    "temperature_heat_index_avg",
    "temperature_heat_index_max",
    "temperature_heat_index_min",
    "temperature_realfeel_avg",
    "temperature_realfeel_max",
    "temperature_realfeel_min",
    "temperature_realfeel_shade_avg",
    "temperature_realfeel_shade_max",
    "temperature_realfeel_shade_min",
    "temperature_wind_chill_avg",
    "temperature_wind_chill_max",
    "temperature_wind_chill_min"
).withColumn(
    "rain_cat",
    when(df.rain_lwe_total > 10, "rainy").otherwise("not rainy")
)

df_selected.write.format("delta").mode("overwrite").saveAsTable("my_weather_delta")

```
